```python
fraud=card[card.Class == 1]
normal=card[card.Class == 0]
```

```python
normal_sample = normal.sample(n=492)
```

```python
new_data = pd.concat([normal_sample, fraud], axis=0)
```

```python
new_data.groupby('Class').mean()
```

```python
x = new_data.drop(columns='Class', axis=1)
y = new_data['Class']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=11)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
train_prediction = model.predict(x_train)
training_accuracy = accuracy_score(train_prediction, y_train)
```

```python
training_accuracy
```

```python
test_prediction = model.predict(x_test)
testing_accuracy = accuracy_score(test_prediction, y_test)
```

```python
testing_accuracy
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

# Predictions
train_prediction = model.predict(x_train)
test_prediction = model.predict(x_test)

# Accuracy calculations
training_accuracy = accuracy_score(y_train, train_prediction)
testing_accuracy = accuracy_score(y_test, test_prediction)

# Plotting the accuracies
labels = ['Training Accuracy', 'Testing Accuracy']
accuracies = [training_accuracy, testing_accuracy]

plt.figure(figsize=(6, 4))
plt.bar(labels, accuracies, color=['skyblue', 'salmon'])
plt.ylim(0, 1)
plt.ylabel('Accuracy')
plt.title('Model Accuracy on Training and Testing Data')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show values on top of bars
for i, acc in enumerate(accuracies):
    plt.text(i, acc + 0.02, f'{acc:.2f}', ha='center')

plt.tight_layout()
plt.show()
```
